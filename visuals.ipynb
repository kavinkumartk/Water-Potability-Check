```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df1 = pd.read_csv(r"C:\Users\kavin\OneDrive\Desktop\End to end\cleaned_water_potability.csv")
df1
```

```python
sns.histplot(df1['ph'],kde=True)
```

```python
df1.drop(columns='Potability').hist(bins=30, figsize=(15, 10), edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
sns.countplot(x='Potability', data=df1)
plt.title("Class Balance: Potability")
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(df1.drop(columns='Potability').columns, 1):
    plt.subplot(3,3,i)
    sns.boxplot(x=df1[col], color="skyblue")
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df1.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
